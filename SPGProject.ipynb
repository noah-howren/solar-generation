```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

solarGen = pd.read_csv('solar.csv', index_col = 'TIMESTAMP', parse_dates = True)

solarGen1 = solarGen[:19703]
solarGen2 = solarGen[19704:39407]
solarGen3 = solarGen[39408:]
```

```python
solarGen1Train = solarGen1[:10942]
solarGen1Test  = solarGen1[10943:]
solarGen1Final = solarGen1Test

solarGen2Train = solarGen2[:10942]
solarGen2Test  = solarGen2[10943:]
solarGen2Final = solarGen2Test

solarGen3Train = solarGen3[:10942]
solarGen3Test  = solarGen3[10943:]
solarGen3Final = solarGen3Test

solarGen1Train.head()
```

```python
from pmdarima import auto_arima
import warnings
warnings.filterwarnings("ignore")

from statsmodels.tsa.seasonal import seasonal_decompose

result_add = seasonal_decompose(solarGen1Train['POWER'], model='additive', extrapolate_trend='freq')

import itertools

p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))

seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]

import statsmodels.api as sm


mod1 = sm.tsa.statespace.SARIMAX(
    solarGen1Test['POWER'],
    order=(1, 1, 1), seasonal_order=(1, 1, 1, 12),
    enforce_stationarity=False,
    enforce_invertibility=False)
results1 = mod1.fit()

pred1 = results1.get_prediction(start=0, dynamic=False, full_results=True)
pred_ci1 =pred1.conf_int()

ax1 = solarGen1Test['POWER'].plot(label='Actual', color='tab:blue')
pred1.predicted_mean.plot(ax=ax1, label='Forecast', color='tab:red', alpha=.5)
plt.ylim(0, 1)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_squared_error  as mse
import math

mae1 = results1.mae
mae2 = results2.mae
mae3 = results3.mae
maeTot = (mae1 + mae2 + mae3)/3

print("Mean Absolute Error - Plant 1 = ", mae1)
print("Mean Absolute Error - Plant 2 = ", mae2)
print("Mean Absolute Error - Plant 3 = ", mae3)
print("Mean Absolute Error - Total   = ", maeTot)

rmse1 = math.sqrt(results1.mse)
rmse2 = math.sqrt(results2.mse)
rmse3 = math.sqrt(results3.mse)
rmseTot = (rmse1 + rmse2 + rmse3)/3

print ("\n")

print("Root Mean Square Error - Plant 1 = ", rmse1)
print("Root Mean Square Error - Plant 2 = ", rmse2)
print("Root Mean Square Error - Plant 3 = ", rmse3)
print("Root Mean Square Error - Total   = ", maeTot)
```
